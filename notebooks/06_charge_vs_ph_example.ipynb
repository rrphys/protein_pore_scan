```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# makes plots pretty
%config InlineBackend.figure_format = 'retina'
```

```python
pH = np.linspace(1, 14, 100)
pK = 6.2

avg_charge = 10**pH / (10**pH + 10**pK)

plt.figure(figsize=(9,6))
plt.plot(pH, avg_charge)
plt.ylabel('Avg. charge of Lysine residue (e)', fontsize=16)
plt.xlabel('Solution pH', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
